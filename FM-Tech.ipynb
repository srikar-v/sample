# Facilities management tech overview

```python
## Libraries
import pymongo
import pandas as pd
from dateutil import parser
from bson.json_util import dumps
import json

from collections import Counter
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Get all the data required
client = pymongo.MongoClient('10.10.0.8', port = 27017)
source_db = client.data_sources.news
```

```python
past_7_days = source_db.find({'tagging_source':'scraped','published_date':{'$gte': parser.parse("2017-10-16T00:00:00.000Z"),
        '$lt':  parser.parse("2017-10-24T00:00:00.000Z")}})
past_7_days = json.loads(dumps(past_7_days))

past_30_days = source_db.find({'tagging_source':'scraped','published_date':{'$gte': parser.parse("2017-09-24T00:00:00.000Z"),
        '$lt':  parser.parse("2017-10-24T00:00:00.000Z")}})
past_30_days = json.loads(dumps(past_30_days))


past_90_days = source_db.find({'tagging_source':'scraped','published_date':{'$gte': parser.parse("2017-07-24T00:00:00.000Z"),
        '$lt':  parser.parse("2017-10-24T00:00:00.000Z")}})
past_90_days = json.loads(dumps(past_90_days))
```

```python
# Get all the social tags
social_tags_7 = []
industry_terms_7 = []
companies_7 = []
for doc in past_7_days:
    social_tags_7.extend(doc['calais_meta']['social_tags'])
    industry_terms_7.extend(doc['calais_meta']['industry_terms'])
    companies_7.extend(doc['calais_meta']['companies'])
    
social_tags_30 = []
industry_terms_30 = []
companies_30 = []
for doc in past_30_days:
    social_tags_30.extend(doc['calais_meta']['social_tags'])
    industry_terms_30.extend(doc['calais_meta']['industry_terms'])
    companies_30.extend(doc['calais_meta']['companies'])
    
social_tags_90 = []
industry_terms_90 = []
companies_90 = []
for doc in past_90_days:
    social_tags_90.extend(doc['calais_meta']['social_tags'])
    industry_terms_90.extend(doc['calais_meta']['industry_terms'])
    companies_90.extend(doc['calais_meta']['companies'])
    
    
```

```python
x = Counter(social_tags_90)
df = pd.DataFrame({'tag':list(x.keys()),'count':[x[key] for key in x.keys()]})
df = df.sort_values('count',ascending = False)
df.head(20)
```

```python
# Get all the companies in internet of things
term = "Emerging technologies"
companies_sub = []
ind_terms_sub = []
social_tags_sub = []
news_sub = []
for doc in past_90_days:
    if term in doc['calais_meta']['social_tags']:
        social_tags_sub.append(doc['calais_meta']['social_tags'])
        ind_terms_sub.append(doc['calais_meta']['industry_terms'])
        companies_sub.append(doc['calais_meta']['companies'])
        news_sub.append(doc['body'])
companies_sub  
```
